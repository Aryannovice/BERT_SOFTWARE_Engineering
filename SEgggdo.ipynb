```python
!pip install numpy==1.26.4 pandas==2.2.2 --force-reinstall



import pandas as pd
import numpy as np
import re
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
from umap import UMAP
from hdbscan import HDBSCAN
from sklearn.feature_extraction.text import CountVectorizer
from gensim.models.coherencemodel import CoherenceModel
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Load your dataset (replace with your actual path)
data_path = "/content/software_requirements_extended.csv"
df = pd.read_csv(data_path)
```

```python
print(df.head())
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df['Cleaned_Requirement'] = df['Requirement'].apply(clean_text)
```

```python
embedding_model = SentenceTransformer("all-mpnet-base-v2")
#all-mpnet-base-v2
embeddings = embedding_model.encode(df['Cleaned_Requirement'], show_progress_bar=True)
```

```python
umap_model = UMAP(n_neighbors=15, n_components=5, metric='cosine', random_state=42)

# Improved HDBSCAN
hdbscan_model = HDBSCAN(min_cluster_size=5, metric='euclidean', cluster_selection_method='eom')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer_model = TfidfVectorizer(ngram_range=(1,3), stop_words="english", max_features=10000)
```

```python
topic_model = BERTopic(
    embedding_model=embedding_model,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    vectorizer_model=vectorizer_model,
    verbose=True
)
```

```python
topics, probabilities = topic_model.fit_transform(df['Requirement'].apply(clean_text))
```

```python
df['Topic'] = topics
```

```python
# Save the results to CSV
output_path = "software_requirements_topics.csv"
df.to_csv(output_path, index=False)
print(f"Topics assigned and results saved to {output_path}")
```

```python
from gensim.models.coherencemodel import CoherenceModel

# Prepare data for coherence calculation: tokenized documents
docs = [req.split() for req in df['Cleaned_Requirement']]
```

```python
topics_bert = topic_model.get_topics()

# Convert BERTopic topics into gensim-compatible format (list of lists of words per topic)
topics_for_coherence = []
for topic_id in sorted(topics_bert.keys()):
    topic_words = [word for word, _ in topics_bert[topic_id]]
    topics_for_coherence.append(topic_words)
```

```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary

# Create a dictionary from your tokenized documents
dictionary = Dictionary([doc.split() for doc in df['Cleaned_Requirement']])

# Get topics from the BERTopic model
topics_bert = topic_model.get_topics()

# Convert BERTopic topics into gensim-compatible format, ensuring token IDs
topics_for_coherence = []
for topic_id in sorted(topics_bert.keys()):
    topic_words = [word for word, _ in topics_bert[topic_id]]
    # Convert words to token IDs using the dictionary, handling missing words
    topic_token_ids = [dictionary.token2id.get(word) for word in topic_words if word in dictionary.token2id]
    # Only add topics with at least one valid token ID to avoid empty topics
    if topic_token_ids:
        topics_for_coherence.append(topic_token_ids)

# Calculate coherence
coherence_model = CoherenceModel(topics=topics_for_coherence,
                                 texts=[doc.split() for doc in df['Cleaned_Requirement']],
                                 dictionary=dictionary,
                                 coherence='c_v')

coherence_score = coherence_model.get_coherence()
print(f"BERTopic Model Coherence Score: {coherence_score:.4f}")
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

# Encode Type column into numeric labels
label_encoder = LabelEncoder()
df['True_Label'] = label_encoder.fit_transform(df['Type'])  # Convert text labels to numbers
```

```python
import numpy as np

# Create a mapping from BERTopic topics to the most frequent true label
topic_label_map = df.groupby('Topic')['True_Label'].agg(lambda x: np.bincount(x).argmax()).to_dict()

# Assign predicted labels based on the most frequent mapping
df['Predicted_Label'] = df['Topic'].map(topic_label_map)
```

```python
print("Classification Report for BERTopic Model:\n")
print(classification_report(df['True_Label'], df['Predicted_Label'], target_names=label_encoder.classes_))
```

```python
print("Classification Report for BERTopic Model:\n")
print(classification_report(df['True_Label'], df['Predicted_Label'], target_names=label_encoder.classes_))
```

```python
from sklearn.metrics import homogeneity_completeness_v_measure, adjusted_mutual_info_score

# Compute homogeneity, completeness, and V-measure
homogeneity, completeness, v_measure = homogeneity_completeness_v_measure(df['Type'], df['Topic'])
ami_score = adjusted_mutual_info_score(df['Type'], df['Topic'])

print(f"Homogeneity: {homogeneity:.4f}")
print(f"Completeness: {completeness:.4f}")
print(f"V-Measure: {v_measure:.4f}")
print(f"Adjusted Mutual Information (AMI): {ami_score:.4f}")
```
